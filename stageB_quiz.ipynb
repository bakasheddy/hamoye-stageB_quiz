```python
cols = {
    'T1':'kitchen-Temp(T1)', 'RH_1': 'kitchen-Humid(RH_1)', 'T2': 'living_room-Temp(T2)', 'RH_2': 'living_room-Humid(RH_2)',
    'T3': 'laundry_room-Temp(T3)', 'RH_3': 'laundry_room-Humod(RH_3)', 'T4': 'office_room-Temp(T4)', 'RH_4': 'office_room-Humid(RH_4)',
    'T5': 'bathroom-Temp(T5)', 'RH_5': 'bathroom-Humid(RH_5)', 'T6': 'outside_north-Temp(T6)', 'RH_6': 'outside_north-Humid(RH_6)',
    'T7': 'ironing_room-Temp(T7)', 'RH_7': 'ironing_room-Temp(RH_7)', 'T8': 'teenager_room_2-Temp(T8)', 'RH_8': 'teenager_room_2-Humid(RH_8)',
    'T9': 'parents_room-Temp(T9)', 'RH_9': 'parents_room-Humid(RH_9)', 'To': 'outside-Temp(Chievres weather station)', 
    'RH_out': 'outside-Humid(Chievres weather station)', 'rv1': 'Random_variable_1(nondimensiona)', 'rv2': 'Random_variable_2(nondimensional)'
}
```

```python
df.rename(columns=cols, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
SCALE = MinMaxScaler()
df = df.drop('date', axis=1)
norm = pd.DataFrame(SCALE.fit_transform(df), columns=df.columns)
```

```python
x = norm.drop(columns=['lights'])
y = norm['Appliances']
```

```python
X_train, X_test, y_train, y_test = train_test_split( x,y, test_size=0.3)
```

```python
print(X_train)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
prediction = lin_model.predict(X_test)
prediction
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, prediction)
round(mae, 2)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, prediction)
round(r2, 3)
```

```python
rss = np.sum(np.square(y_test - prediction))
round(rss, 2)
```
